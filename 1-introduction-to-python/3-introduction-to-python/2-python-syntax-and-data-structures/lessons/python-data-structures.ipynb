# Python Data Structures

Now it's time to take a closer look at four **built-in data structures** in Python. Each one is associated with a type of brackets:
(A lot of what you have learned so far is recyclable here!)

| Name | Description   | Brackets   |
|:- |:- |:- |
|   **lists**  | an editable(mutable), **ordered collection** of values |Associated with square brackets `[]`|
|   **dictionaries**  | an unordered, editable collection of **`key:value` pairs** |Associated with curly brackets `{}`|
|   **tuples**  | similar to lists, but **not editable**(immutable) |Associated with round brackets `()`|
|   **sets**  | an unordered, editable collection **unique values**|Associated with curly brackets `{}`|

- Data structures are used to **store** and **organise** data

- You are likely to encounter **lists** and **dictionaries** more frequently than **tuples** and **sets**  

- You will use other, more **specialised data structures** in future when using **packages** such as **pandas**

## ☑️ Lists
In Python a list is created by placing all the elements inside **square brackets [ ]**, separated by commas

- Lists have  a variable number of elements
- Elements can be **added**, **removed** or **modified**
- Elements are accessed using a **zero-based index**
- Elements do not have to be the same data type
- You can also define an empty list with square brackets, e.g. `my_list = []`
- You can also create a list using the built-in `list()` function

#### List indexing

The syntax we can use for **accessing list elements by position** are the same as we have seen previously for strings, with the number of **numerical values** and **colons** determining the interpretation:

- `[position]`
- `[start:stop]`
- `[start:stop:step]`

Again, indexing is **zero-based**, and **negative values** can be used to access by **position from the end of the list**.

#### Define a list using square brackets `[]`

```python
rain = [6.5, 0, 0, 1.2, 2.6, 5.4]
```

```python
rain
```

```python
type(rain)
```

```python
len(rain)
```

- The built-in `len()` function can be used on lists and strings to determine their **length**

#### List indexing
- Retrieving  items from the list, using indexing (remember that Python is zero indexed)

```python
rain[4]
```

```python
rain[1:4]
```

```python
rain[4:]
```

```python
rain[:4]
```

#### List indexing with negative values

```python
rain[-2]
```

```python
rain[-2:]
```

```python
rain[:-2]
```

#### List indexing with step values

```python
rain[::2]
```

```python
rain[::-2]
```

```python
rain[::-1]    #We can use negative values for `step` to return values in **reverse order**
```

## ☑️  Dictionaries
You can define a dictionary by using curly brackets `{key:value}`, with each key and value separated by a colon `:` 

- Dictionaries **map values to keys** `{key:value}`
- Each item stored in a dictionary is actually a **key:value pair**
- Values are **accessed using the key** (rather than index)
- Values can be modified and further **key:value pairs** added
- You can also define an empty dictionary with curly brackets, e.g. `my_dict = {}`
- You can also create a dictionary using the built-in `dict()` function

#### Define a dictionary using curly brackets `{key:value}`

```python
user = {'Name': 'David',
        'Occupation': 'Magician',
        'Phone': 5554267,
        'Town': 'New York',
       }
```

```python
user
```

```python
type(user)
```

```python
len(user)
```

We can use `keys` inside square brackets `[ ]` to **retrieve** `values`

```python
user['Town']
```

```python
user['Phone']
```

We can **add** or **modify** values in the same way as we would assign or update variables:

```python
user['Town'] = 'San Francisco'    #modify
user['Gender'] = 'Male'           #add
```

```python
user
```

#### Python Dictionary Methods
- Python has a set of built-in methods that are available with every dictionary object

| Method | Description   |
|:- |:- |
|copy()| creates a copy of the current dictionary|
|get()|Returns the value of the specified key|
|items()|Returns a list containing a tuple for each key-value pair|
|keys()|Returns a list containing dictionary keys|
|pop()|Removes items with the specified key|
|update()|Updates the dictionary with the specified `{key:value}` pairs|
|values()|Returns a list of all the values in the dictionary|

#### Some useful references:

[Python Dictionary Methods](https://docs.python.org/3/tutorial/datastructures.html#dictionaries)

[Python Dictionary Methods](https://www.w3schools.com/python/python_dictionaries_methods.asp)

```python
user
```

```python
user.update({'Town': 'Washington',
             'Country': 'USA'})
user
```

- `.update()` allows us to **update** an existing dictionary using another dictionary of `key:value` pairs

```python
user['Age']
```

```python
user.get('Age', 'Not provided')
```

 - `.get()` allows us to provide a **fallback value** should the given key not be found in the dictionary

```python
user.keys()
```

```python
user.values()
```

`.keys()` and `.values()` return a *view* of the **keys** and **values** respectively; we can use the **built-in `list()` function** to return the list itself from each:

```python
list(user.keys())
```

```python
list(user.values())
```

## ☑️ Tuples

In Python a tuple is created by placing all the elements inside **round brackets ( )**, separated by commas

- Tuples contain an **ordered collection** of elements
- The **number** of elements are **fixed** 
- Individual elements **cannot be modified** - immutable
- Elements are accessed using a **zero-based index**
- Elements do not have to be the same data type
- You can also create a tuple using the built-in `tuple()` function

#### Define a tuple using round brackets `( )`

```python
british_values = ('Democracy',
                  'Rule of law',
                  'Individual liberty',
                  'Mutual respect',
                  'Tolerance of those of different faiths and beliefs')
```

```python
british_values
```

```python
type(british_values)
```

```python
len(british_values)
```

```python
british_values[2]
```

##### Tuples vs lists

The key difference between tuples and lists is that `tuples` are immutable and `lists` are mutable. In practice, you can achieve the same functionality by using a `list`.
 

## ☑️  Sets

In Python a set is created by placing all the elements inside **curly braces { }**, separated by commas

- All elements in a set are **unique** (no duplicates)
- Elements are **not ordered or indexed**
- A set can contain any number of elements with different data types (integer, float, string, bool)
- It is also possible to use the `set()` function to make a set, this is useful when converting a `list` or `tuple` to a `set`


#### Practical Use Cases of Python Sets
Because sets cannot have multiple occurrences of the same element, sets are highly useful while removing duplicate values from a list or tuple.


#### Define a set using curly brackets `{ }`

```python
friends = {'Ellie', 'Sarah', 'Amar', 'Sarah', 'Sarah', 'Sarah'}
friends
```

```python
type(friends)
```

```python
len(friends)
```

#### Python Set Methods
- Python has a set of built-in methods that are available with every set object

| Method | Description   |
|:- |:- |
|copy()|Returns a copy of the set|
|discard()|Removes the specified item from the set|
|update()|Update the set with the union of this set and others|


#### Some useful references:

[Python Set Methods](https://docs.python.org/3/tutorial/datastructures.html#sets)

[Python Set Methods](https://www.w3schools.com/python/python_ref_set.asp)

```python
friends.update(['James', 'Ellie'])
friends
```

```python
friends.discard('Sarah')
friends
```

```python
sorted(friends, reverse=True)
```

Notice how the original **order is not retained**. Although here the items are displayed alphabetically, the order in which set items are stored in memory should be considered **arbitrary**.

```python
required = ['F', 'A', 'C', 'B', 'A', 'C', 'A', 'D', 'H', 'B']   #define a list
required
```

```python
set(required)
```

 - The built-in `set()` function is useful for identifying the **unique values** in a list

```python
list(set(required))
```

```python
sorted(list(set(required)), reverse=True)
```

- We can use the built-in `list()` function to return a list
- As mentioned previously, the item order of a set is arbitary, but can be resolved using the built-in `sorted()` function

We can check for **membership** of a set as follows:

```python
'Sarah' in friends
```

```python
"Dan" in friends
```

The same `in` keyword can also be used to check for the **presence of a value** in a `list`:

```python
rain = [6.5, 0, 0, 1.2, 2.6, 1.9, 5.4]
rain
```

```python
6.5 in rain
```

```python
0 in rain
```
