## Functions

A **function** performs a **specified operation** using one or more **inputs**.  

- The result of a function operation is typically the **output** of a new object, or the **modification** of an existing object.  

- The types of inputs required or accepted by a function are the **parameters**.  

- The specific inputs provided when a function is **called** on a given occasion are known as **arguments**.

Let's consider a simple example of a **mathematical function**:

```
   y = 2x + 3
```
In maths this could alternatively be written in **function notation**, such as:
```
f(x) = 2x + 3
```

The purpose of this is to allow us to see more clearly the result of the given function when a particular value is used for `x`:
```
f(3) = 2(3) + 3
f(3) = 9
```

In Python, this function could be represented as follows:

````python
def my_func(x):
    
    y = (2 * x) + 3
    
    return y
````

````python
my_func(5)
````

Both `x` and `y` are arbitrary strings:
- Having been used as the **parameter name**, `x` is used within the **function definition** to refer to the argument provided for `x` when the function is called, in this case `5`

- The result of the calculation is assigned to `y` within the function, and subsequently referenced in the `return` statement

`x` and `y` are only assigned **temporarily**, i.e. only within and for the purposes of the execution of the function on any given occasion.

### Syntax 
```python
def functionname( parameters ):
    "function_docstring" 
    function_suite
    return expression
```
- `def` is the Python keyword to indicate that we are about to **define** a function
- `my_func` is the **name** of our function, as per variables it **cannot contain spaces**
- `x` is the only **parameter** for this particular function
- parentheses ( `( )` ) which may or may not include parameters; parameters can be assigned inside these parenthese
- and a colon (`:`)

### `functions` 

````python
def times_two(x):
    """Multiplies the 
    argument by two."""
    result = 2*x
    return result

a = 2
b = 3
c = 4
print(times_two(a))
print(times_two(b))
print(times_two(c))
````

````python
my_var = times_two(8)
my_var
````

The function can also handle non-numerical values:

````python
print(times_two('a'))
````

````python
print(times_two([1, 3]))
````

Let's get the documentation for this function:

````python
# This will show the documentation in line
help(times_two)
````

````python
# This will show the documentation in a separate window
?times_two
````

````python
# You can do this for any function
help(len)
````

Functions are not only useful for mathematical operations.
As an example, let's create a function called **printme**.

````python
def printme(my_string):
    """
    This prints a passed string into this function
    """
    print(my_string)
    return my_string
````

Now, let's call the function:

````python
printme("M25 traffic makes my head hurt.")
printme("It looks like a car parks out there!")
````

````python
a = printme("M25 traffic makes my head hurt.")
````

````python
# If you don't use return in the function, nothing will be assigned to the variable.
a
````

````python
type(a)
````

Functions can have additional keyword arguments to define default values. Compare the output.

````python
def times_two_key(x=2):
    """Multiplies the 
    argument by two."""
    return 2*x
````

````python
print(times_two_key())
print(times_two_key(3))
````

We can also mix both kinds of arguments (the order is important, keyword arguments come at the end).

````python
def multiply_and_add(x, y=1, z=2):
    """Returns the product and the sum of the arguments."""
    return x*y*z, x+y+z
````

````python
print(multiply_and_add(3))
print(multiply_and_add(3, 5, 4))
print(multiply_and_add(3, z=4, y=4))
````

### `functions` 

````python
helen_tips = [0.50, 1.25, 2.45, 0.37, 1.20, 3.70, 5.00]

def average_tip(tips):
    
    avg = sum(tips) / len(tips)
    
    return avg

average_tip(helen_tips)
````

Having defined the `average_tip` function, we can now **call** the `average_tip` function with any suitable **argument** for the `tips` parameter:

````python
joe_tips = [0.40, 3.87, 2.60, 1.00, 3.40]

joe_avg = average_tip(joe_tips)
joe_avg
````

Python will return an error if the input cannot be processed:

````python
average_tip(helen_tips, joe_tips)
````

Reading **error messages** will help you to understand *why* things went wrong.

### Built-in functions

Python has a number of **built-in functions**, which are always available. In fact we used a couple of these within the function we defined previously:

- `len()`: returns the **length** (or number of items) in an object
- `sum()`: returns the **sum** total of the numbers in an object

As with our own function, these functions will only work when a compatible **argument** is **passed** (input) to it. 

## Methods

A **method** is a **function** which is **associated with a given object**.

- a function performs an operation on one or more objects provided to it, e.g. `print()`
- a method performs an operation on a specific object, e.g. `.format()`

Objects of a particular **type** (such as lists or dictionaries) will have the **same methods** available.

Some examples of methods on lists:


````python
helen_tips.append(3.50)
print(helen_tips)
````

````python
helen_tips.sort()
print(helen_tips)
````
