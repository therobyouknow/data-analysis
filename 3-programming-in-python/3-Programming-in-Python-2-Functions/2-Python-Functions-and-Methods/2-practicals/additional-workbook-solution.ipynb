**Question 15** A Fibonacci Sequence looks like this: 1, 1, 2, 3, 5, 8, 13, etc., ie each number is the sum of the preceding two.

Write a while loop which creates a list which contains the Fibonacci Sequence up to a maximum of 50.

Your resulting list should be assigned variable `list_fib`.

````python
#add your code below

list_fib = [1, 1]
x = 1
y = 1
while x + y < 50:
    z = x + y
    list_fib.append(z)
    x = y
    y = z
list_fib

````

**Question 16** Fizzbuzz: have you ever played fizzbuzz, a game normally played in a group? Each player needs to count incrementally through integers, replacing any number divisible by 3 with the word 'fizz' and any number divisible by 5 with 'buzz'. If the number is divisible by both 3 and 5 then it is replaced by 'fizzbuzz'.

Write code that creates a list of numbers which follow the rules of fizzbuzz, from 1 to 34 inclusive, and assign the result to variable `list_fb`.

````python
#add your code below

list_fb = []
for i in range(1, 35):
    element = ''
    if i % 3 == 0:
        element = element + 'fizz'
    if i % 5 == 0:
        element = element + 'buzz'
    if len(element) == 0:
        element = i
    list_fb.append(element)

list_fb

````

**Question 22** The rules for the password for a particular website are:

- Must be at least 8 characters long
- Must contain a mix of upper case, lower case and numbers

Write a function which takes `password` and returns `True` if it satisfies these conditions, and `False` otherswise. Assign the result to variable `passed`.

````python
password = 'PYT12hon'
````

````python
#add your code below

def pwd_check(password):
    u, l, n = 0, 0, 0
    
    if len(password) >= 8:
        for char in password:
            if char.isupper():
                u += 1
            elif char.islower():
                l += 1
            elif char.isdigit():
                n += 1
    if (u >= 1 and l >= 1 and n >= 1):
        result = True
    else:
        result = False
    return result

passed = pwd_check(password)
passed

````
